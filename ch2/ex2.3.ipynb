#### 1. Create and store a sequence of values from 5 to −11 that progresses in steps of 0.3.

```r
seq_obj = seq(from=5, to=-11, by=-0.3)  # this is a decreasing sequence, so negative step
seq_obj
```

#### 2. Reverse the order of the object above.

```r
seq_obj = sort(seq_obj, decreasing=FALSE)
seq_obj
```

#### 3. Repeat the vector c(-1,3,-5,7,-9) twice, with each element repeated 10 times, and store the result. Display the result sorted from largest to smallest.

```r
new_obj = rep(c(-1, 3, -5, 7, -9), times=2, each=10)
sort(new_obj, decreasing=TRUE)
```
